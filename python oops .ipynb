## Object-Oriented Programming (OOP) in Python

Object-Oriented Programming (OOP) is a programming paradigm based on the concept of 'objects', which can contain data and code. The main goal of OOP is to increase the flexibility and maintainability of programs. Python is a multi-paradigm language, and it fully supports OOP.

### Core Principles of OOP:

1.  **Encapsulation:** Bundling data (attributes) and methods (functions) that operate on the data into a single unit or class. It also restricts direct access to some of an object's components, which means direct access to the state of an object is prevented and only allowed through an object's methods.
2.  **Inheritance:** A mechanism where a new class (subclass/child class) is derived from an existing class (superclass/parent class), inheriting its attributes and methods. This promotes code reusability.
3.  **Polymorphism:** The ability of an object to take on many forms. In Python, this means methods can be defined in a parent class and then overridden in child classes. It also refers to the ability of different classes to respond to the same method call in their own specific way.
4.  **Abstraction:** Hiding the complex implementation details and showing only the essential features of the object. In Python, this can be achieved using abstract classes and methods.

### 1. Classes and Objects

-   **Class:** A blueprint or a template for creating objects. It defines a set of attributes and methods that the created objects will have.
-   **Object:** An instance of a class. When a class is defined, no memory is allocated. Memory is allocated only when an object (an instance) of the class is created.

Let's define a simple `Dog` class.

```python
class Dog:
    # Class attribute
    species = "Canis familiaris"

    def __init__(self, name, age):
        # Instance attributes
        self.name = name
        self.age = age

    def bark(self):
        return f"{self.name} says Woof!"

    def description(self):
        return f"{self.name} is {self.age} years old."

# Creating objects (instances) of the Dog class
my_dog = Dog("Buddy", 3)
your_dog = Dog("Lucy", 5)

print(f"My dog's name: {my_dog.name}")
print(f"Your dog's age: {your_dog.age}")
print(f"My dog's species: {my_dog.species}")
print(my_dog.bark())
print(your_dog.description())
```

### 2. Inheritance

Inheritance allows us to define a class that inherits all the methods and properties from another class. The original class is called the parent class (or base class), and the new class is called the child class (or derived class).

Let's create a `GoldenRetriever` class that inherits from `Dog`.

```python
class GoldenRetriever(Dog):
    def __init__(self, name, age, color):
        super().__init__(name, age) # Call the parent class's constructor
        self.color = color

    def fetch(self, item):
        return f"{self.name} is fetching the {item}!"

    # Method Overriding: GoldenRetriever's bark is different
    def bark(self):
        return f"{self.name} says Yap! Yap!"

my_golden = GoldenRetriever("Max", 2, "Golden")

print(f"Golden Retriever's name: {my_golden.name}")
print(f"Golden Retriever's age: {my_golden.age}")
print(f"Golden Retriever's color: {my_golden.color}")
print(f"Golden Retriever's species: {my_golden.species}") # Inherited class attribute
print(my_golden.description()) # Inherited method
print(my_golden.fetch("ball")) # New method
print(my_golden.bark()) # Overridden method
```

### 3. Polymorphism

Polymorphism means 'many forms'. It allows us to define methods in the child class that have the same name as the methods in the parent class. In polymorphism, Python will use the method of the object that is calling it.

Here, both `Dog` and `GoldenRetriever` objects can respond to the `bark()` method, but in their own way.

```python
def make_animal_bark(animal):
    print(animal.bark())

make_animal_bark(my_dog) # Calls Dog's bark()
make_animal_bark(my_golden) # Calls GoldenRetriever's bark()
```

### 4. Encapsulation

Encapsulation is about bundling the data (attributes) and methods (functions) that operate on the data into a single unit (class). It also involves restricting direct access to some of an object's components. In Python, encapsulation is achieved by convention, using single or double underscores to denote 'protected' or 'private' attributes/methods.

-   **Protected members (single underscore `_`):** Intended for internal use within the class or its subclasses, but can still be accessed externally.
-   **Private members (double underscore `__`):** Python internally mangles these names, making them harder (but not impossible) to access directly from outside the class. This is mainly to avoid name clashes in inheritance.

These are the fundamental concepts of OOP in Python. Mastering them will help you write more organized, reusable, and maintainable code.

### Basic NumPy Array Example

NumPy (Numerical Python) is a fundamental package for scientific computing in Python. It provides an array object that is up to 50x faster than traditional Python lists.

This example shows:
1.  Importing the NumPy library.
2.  Creating a 1-dimensional array (vector).
3.  Performing an element-wise operation (adding a scalar value).

```python
import numpy as np

# 1. Create a NumPy array from a Python list
my_array = np.array([1, 2, 3, 4, 5])
print(f"Original array: {my_array}")
print(f"Type of array: {type(my_array)}")
print(f"Shape of array: {my_array.shape}") # (5,) indicates a 1D array with 5 elements

# 2. Perform an element-wise operation (add 10 to each element)
new_array = my_array + 10
print(f"Array after adding 10: {new_array}")

# 3. Create a 2-dimensional array (matrix)
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(f"\nOriginal matrix:\n{matrix}")
print(f"Shape of matrix: {matrix.shape}") # (2, 3) indicates 2 rows and 3 columns

# 4. Perform another element-wise operation (multiply by 2)
multiplied_matrix = matrix * 2
print(f"Matrix after multiplying by 2:\n{multiplied_matrix}")
```

### Data Analysis with Pandas

Pandas provides a rich set of tools for performing various data analysis tasks. Here are some common operations:

1.  **Selection and Filtering:** Extracting specific rows or columns based on conditions.
2.  **Aggregation:** Grouping data and applying statistical functions (e.g., sum, mean, count) to these groups.
3.  **Sorting:** Arranging data by one or more columns.
4.  **Handling Missing Data:** Identifying and dealing with `NaN` (Not a Number) values.
5.  **Applying Functions:** Using `apply()`, `map()`, or `applymap()` for custom transformations.
6.  **Merging and Joining:** Combining DataFrames based on common keys.
7.  **Pivot Tables:** Reshaping data for summarization and analysis.

Let's demonstrate a few of these with our existing DataFrame `df`.

```python
# Assuming 'df' from the previous example is available

print("\n--- Data Analysis Operations ---")

# 1. Selection and Filtering (already shown, but let's do another one)
print("\nPeople in New York:")
print(df[df['City'] == 'New York'])

# 2. Aggregation: Group by City and calculate average Age
print("\nAverage Age by City:")
print(df.groupby('City')['Age'].mean())

# 3. Sorting: Sort by Age in descending order
print("\nDataFrame sorted by Age (descending):")
print(df.sort_values(by='Age', ascending=False))

# 4. Handling Missing Data (demonstration with a copy)
df_with_nan = df.copy()
df_with_nan.loc[0, 'Occupation'] = None # Introduce a missing value
df_with_nan.loc[2, 'Age'] = np.nan # Introduce another missing value

print("\nDataFrame with missing values:")
print(df_with_nan)

print("\nChecking for missing values:")
print(df_with_nan.isnull().sum())

print("\nDataFrame after dropping rows with any missing values:")
print(df_with_nan.dropna())

print("\nDataFrame after filling missing 'Occupation' with 'Unknown':")
# Fill only 'Occupation' column, original df_with_nan is modified
print(df_with_nan['Occupation'].fillna('Unknown'))

# 5. Applying Functions: Create a new column based on a condition
df['Age_Group'] = df['Age'].apply(lambda x: 'Adult' if x >= 18 else 'Minor')
print("\nDataFrame with new 'Age_Group' column:")
print(df)
```

### Introduction to Pandas

Pandas is an open-source Python library widely used for data manipulation and analysis. It provides powerful, flexible, and easy-to-use data structures, most notably the `DataFrame` and `Series`.

Key features of Pandas:
*   **DataFrame:** A 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dictionary of Series objects.
*   **Series:** A 1-dimensional labeled array capable of holding any data type.
*   **Data Alignment:** Data is automatically aligned by labels, which can prevent common errors and make your code more robust.
*   **Powerful Data Operations:** Offers a rich set of functionalities for data cleaning, transformation, aggregation, merging, and more.
*   **Handling Missing Data:** Provides tools for easily dealing with missing data (`NaN`).

```python
import pandas as pd

# 1. Create a Series (1-dimensional labeled array)
ages = pd.Series([22, 35, 58, 41], name='Age')
print("\n--- Pandas Series ---")
print(ages)
print(f"Type of Series: {type(ages)}")

# 2. Create a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)

print("\n--- Pandas DataFrame ---")
print(df)
print(f"Type of DataFrame: {type(df)}")
print(f"DataFrame shape: {df.shape}")

# 3. Accessing columns
print("\n--- Accessing a column (Series) ---")
print(df['Name'])

# 4. Accessing rows by index
print("\n--- Accessing a row by index (iloc) ---")
print(df.iloc[1]) # Second row (index 1)

# 5. Basic filtering
print("\n--- Filtering data (Age > 30) ---")
print(df[df['Age'] > 30])

# 6. Adding a new column
df['Occupation'] = ['Engineer', 'Artist', 'Doctor', 'Scientist']
print("\n--- DataFrame after adding 'Occupation' column ---")
print(df)

# 7. Descriptive statistics
print("\n--- Descriptive statistics for numerical columns ---")
print(df.describe())
```
